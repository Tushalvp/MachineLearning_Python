## 5. Develope Predictive Model

We will use **Mean Squared Error (MSE)** as our metric to evaluate model performance

### 6 Feature Engineering   

**Insights we gathered from Data Exploration**:

- There is a `weak positive` relationship (0.38) between **salary** and **yearsExperience**. 

- There is a `weak negative` relationship between (-0.3) **salary** and **milesFromMetropolis**.

The prediction will be unreliable with weak correaltion. We will engineer new features to enhance model performance.

**`New Features`**:Calculate descriptive statistics by aggregating categorical features (Eg: `Group_mean`, `Group_min`, `Group_max`, `Group_std`)

```python
fillna=True

# Create New Features:
GrpAvg_ds, GrpAvg_cols = FeatureEng_GrpAvg(train_df, cat_cols, target_col)

# Merge new features with existing data
train_dataset = FeatureEng_merge(train_df, GrpAvg_ds, cat_cols, GrpAvg_cols, fillna)
test_dataset = FeatureEng_merge(test_df, GrpAvg_ds, cat_cols, GrpAvg_cols, fillna)
```

```python
# One hot Encoding

Label_encoders={}
for col in cat_cols:
    train_dataset[col], test_dataset[col], Label_encoders[col] = LabelEncode(train_dataset, test_dataset, col)
```

```python
# Updating Feature list

cat_cols_FE = cat_cols + GrpAvg_cols
feature_cols = cat_cols_FE + num_cols
```

```python
#Shuffling training dataset

train_dataset = shuffle_df(train_dataset)
```

```python
# initialise models list and metrics
models = []
mean_mse = {}
cv_std = {}
res = {}

# defining the number of processors
n_procs = 4

# set verbose level
verbose_lvl = 0
```

```python
# creating models
lr = LinearRegression()

rf = RandomForestRegressor(n_estimators = 80, n_jobs = n_procs, max_depth = 20, min_samples_split = 70, \
                          max_features = 10, verbose = verbose_lvl)

gbm = GradientBoostingRegressor(n_estimators = 40, max_depth = 7, loss = 'ls', verbose = verbose_lvl)

models.extend([lr, rf, gbm])
```

### CrossValidation

```python
print('Cross validating the models')
for model in models:
    cross_val_model(model, train_dataset[feature_cols], train_dataset[target_col],n_procs ,mean_mse, cv_std)
    print_summary(model, mean_mse, cv_std)
```

**Mean Squared Error of Models are:**
<br>    Gradient Boosting Regressor : 313.09
<br>    Random Forest Regressor : 314.55
<br>    Linear Regression : 358.17


        

```python
# selecting the best model based on MSE
bestModel = min(mean_mse, key = mean_mse.get)
print('The best model based on MSE during cross validation:')
print(bestModel)
```

```python
bestModel.fit(train_dataset[feature_cols], train_dataset[target_col])
```

```python
predicted_salaries = bestModel.predict(test_dataset[feature_cols])
```

```python
# Writing the test dataset to the csv file
test_dataset['predicted_salary'] = predicted_salaries.tolist()
test_dataset.to_csv(results_path, index = False)
```

```python
feature_importances = get_model_feature_importances(bestModel, train_dataset[feature_cols])
```
